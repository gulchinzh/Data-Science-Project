## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
# default sheet_name is 0. It means the first sheet comes into.

pd.read_excel("Excel_Sample.xlsx")
```

```python
pd.read_excel("Excel_Sample.xlsx", sheet_name = 'Sheet2')
```

```python
#the empty cells of exel file appear as "NaN" 
```

```python
df2 = pd.read_excel("Excel_Sample.xlsx", sheet_name = 'Sheet2')
df2
```

### Excel Output

```python
df2.to_excel("df2_sample.xlsx", sheet_name='Terry', index=False)
```
