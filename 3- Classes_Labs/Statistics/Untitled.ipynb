```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = x*2
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**2
b
```

```python
fig = plt.figure()

axes2 = fig.add_axes([0,0,1,1])
axes2.plot(a,b)

axes1 = fig.add_axes([0.2,0.4,0.3,0.2])
axes1.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)
axes[1].plot(a,b,color = "red")
```

```python
fig,axes = plt.subplots(1, 2)
```
