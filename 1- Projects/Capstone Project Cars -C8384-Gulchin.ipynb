```python
df1.Fuel.value_counts
```

```python
def Fuel_type(x):
    if 'Benzine' in x or 'Gasoline' in x or 'Super' in x:
        return 'Gasoline'
    elif 'Diesel' in x:
        return 'Diesel'
    elif 'LPG' in x:
        return 'LPG'
    else:
        return 'Others'
```

```python
df1.Fuel = df1.Fuel.apply(Fuel_type)
df1.Fuel.value_counts()
```
