#### Body_Type

```python
df1.groupby(['Make','Model']).Body_Type.value_counts(dropna = False)
```

<br>

```python
list(df1[df1.Body_Type.isnull()]['Model'].unique())
```

```python
null_models = list(df1[df1.Body_Type.isnull()]['Model'].unique())
for i in null_models:
    null_model_indexes = df1[(df1.Model == i) & (df1.Body_Type.isnull())].Body_Type.index
    my_mode = df1[(df1.Model == i)].Body_Type.mode()
    df1.Body_Type.loc[null_model_indexes]= df1.Body_Type.loc[null_model_indexes].fillna(my_mode[0])
    print(df1[df1.Body_Type.isnull()].value_counts())  
```

#### Displacement

```python
df1.Displacement
```

```python
def fill_with_mode(a):
    column_name = a
    null_models = list(df1[df1[column_name].isnull()].Model.unique())
    for i in null_models:
        null_model_indexes = df1[(df1.Model == i) & (df1[column_name].isnull())][column_name].index
        my_mode = list(df1[(df1.Model == i)][column_name].mode())
        if my_mode != []:
            df1[column_name].loc[null_model_indexes]= df1[column_name].loc[null_model_indexes].fillna(my_mode[0])

fill_with_mode('Displacement')
```
